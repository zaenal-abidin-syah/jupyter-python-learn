```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value
from jcopml.feature_importance import mean_score_decrease
```

# Import data

```python
df = pd.read_csv("data/creditcard.csv")
df.head()
```

```python
plot_missing_value(df)
```

# Dataset Spliting

```python
X = df.drop(columns="Class")
y = df.Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Training

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from jcopml.tuning import grid_search_params as gsp
```

```python
X_train.columns
```

```python
preprocessor = ColumnTransformer([
    ("numeric", num_pipe(scaling="standard"), X_train.columns)
])
pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo", SVC(max_iter=500))
])

model = GridSearchCV(pipeline, gsp.svm_params, cv=3,n_jobs=-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

99 % ingat setiap kali kita mendapatkan score setinggi itu, kita harus curiga dan cek kembali

- apakah ada data leakage
- apakah datasetnya imbalance
- apakah memang problemnya yang sederhana

pada kasus ini datasetnya imbalance

masalahnya adalah menggunakan akurasi sebagai metrics evaluasi yang mana akurasi tidak cocok untuk dataset yang imbalance, ada scoring yaitu F1-score yang cocok digunakan untuk dataset imbalance
note : jika ingin menggunakan akurasi gunakan resampling / balancing data

# Akurasi tidak bisa dipakai untuk imbalance dataset

By default, kita tidak boleh pakai akurasi sebagai metrik ketika datanya imbalance, Pakailah F1-score. 

```python
preprocessor = ColumnTransformer([
    ("numeric", num_pipe(scaling="standard"), X_train.columns)
])
pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo", SVC(max_iter=500))
])

model = GridSearchCV(pipeline, gsp.svm_params,scoring="f1", cv=3,n_jobs=-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

```python
from jcopml.plot import plot_confusion_matrix
```

```python
plot_confusion_matrix(X_train, y_train, X_test, y_test, model)
```

# Improve Menggunakan Class Weight

```python
# [
#     {0: .05, 1:.95},
#     {0: .1, 1:.9},
#     {0: .25, 1:.75}
# ]
[{0:x, 1:1-x}  for x in [.05, .1, .25]]
```

```python
preprocessor = ColumnTransformer([
    ("numeric", num_pipe(scaling="standard"), X_train.columns)
])
pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo", SVC(max_iter=500))
])

parameter = {
    'algo__gamma': np.logspace(-3,3,7),
    'algo__C': np.logspace(-3,3,7),
    'algo__class_weight' : [{0:x, 1:1-x}  for x in [.01,.05, .1]]
}

model = GridSearchCV(pipeline, parameter,scoring="f1", cv=3,n_jobs=-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

```python
plot_confusion_matrix(X_train, y_train, X_test, y_test, model)
```
