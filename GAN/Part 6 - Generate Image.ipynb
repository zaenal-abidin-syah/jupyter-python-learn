```python
import torch
from torchvision.utils import save_image
import os
from IPython.display import Image
os.makedirs("output", exist_ok=True)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

# Generate Fake Images

### GAN

```python
model="GAN"
G = torch.load(f"model/{model}/generator.pth", map_location="cuda:0").eval()
fake_img = G.generate(64, device)
save_image(fake_img.view(-1, 1, 28, 28), f"output/{model}.jpg", nrow=8, normalize=True)
Image(f"output/{model}.jpg")
```

### CGAN

```python
model = "CGAN"
G = torch.load(f"model/{model}/generator.pth", map_location="cuda:0").eval()
labels = torch.ones(64, device=device).to(int) * 1
fake_img = G.generate(labels, device)
save_image(fake_img.view(-1, 1, 28, 28), f"output/{model}.jpg", nrow=8, normalize=True)
Image(f"output/{model}.jpg")
```

### WGAN

```python
model="WGAN"
G = torch.load(f"model/{model}/generator.pth", map_location="cuda:0").eval()
fake_img = G.generate(64, device)
save_image(fake_img.view(-1, 1, 28, 28), f"output/{model}.jpg", nrow=8, normalize=True)
Image(f"output/{model}.jpg")
```

### WDCGAN

```python
model="WDCGAN"
G = torch.load(f"model/{model}/generator.pth", map_location="cuda:0").eval()
fake_img = G.generate(64, device)
save_image(fake_img, f"output/{model}.jpg", nrow=8, normalize=True)
Image(f"output/{model}.jpg")
```
