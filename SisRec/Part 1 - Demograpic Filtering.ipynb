# Three most common Recommender system

- Demographic filtering
  - rekomendasi yang general (tidak sepenuhnya personalize)
  - filter hanya berdasarkan fitur demographic seperti genre, durasi dll
  - the simplest recommender system
  - ide : rekomendasi yang disukai secara umum
- Content-based Filtering
  - rekomendasi content/barang yang serupa
  - filter menggunakan fitur yang lebih spesifik seperti genre, sinopsis dll.
  - jika seseorang menonton X maka dia akan direkomendasikan film yang mirip x
- Collaborative Filtering
  - mencocokan orang dengan preferensi yang serupa
  - tidak membutuhkan filter data apapun, hanya perlu mencari similarity dengan yang lain

# Simple Demographic Filtering : Filter -> Scoring -> Sort

```python
df = pd.read_csv('data/demographic.csv')
df.head()
```

```python
df.sort_values('vote_average', ascending=False)
```

# Step 1 : Filter

```python
genres = ['Animation']
duration = (60, 150)
year = (2000, 2019)
topk = 20
```

```python
df = df[
    df.release_year.between(year[0], year[1]) &
    df.runtime.between(duration[0], duration[1]) &
    df[genres].all(axis=1) 
]
df.head()
```

# Step 2 : Scoring

```python
recommendation = df.loc[:, "title":"release_year"]
recommendation = recommendation.sort_values('vote_average', ascending=False).head(topk)
recommendation
```

# IMDB Weighted Rating

Film dengan 100 orang yang vote 9.5 seakan-akan lebih baik daripada 10000 orang dengan vote 9.0. <br>
harusnya dibobot dengan jumlah orang yang vote juga. <br>IMDB Weighted Rating :
\begin{equation}
WR = \frac{Rv+Cm}{v+m}
\end{equation}

- $v$ : jumlah voting film -> vote_count
- $m$ : syarat minimum vote -> menggunakan quantile
- $R$ : rata-rata rating -> vote_average
- $C$ : rata-rata rating semua film -> bisa dihitung

```python
def imdb_score(df):
    df = df.copy()
    m = df.vote_count.quantile(0.85)
    C = (df.vote_average * df.vote_count).sum() / df.vote_count.sum()
    df = df[df.vote_count >= m]
    df['score'] = df.apply( lambda x: (x.vote_average * x.vote_count + C * m) / (x.vote_count + m), axis=1)
    return df
                    
```

```python
df = imdb_score(df)
```

```python
recommendation = df.sort_values('score', ascending=False).head(topk)
recommendation = recommendation.loc[:, "title":"release_year"]
recommendation
```

# ML Enginering : Sum Them All Up

```python
class RecommenderSystem:
    def __init__(self, data):
        self.df = pd.read_csv(data)
    def recommend(self, genres=None, duration=None, year=None, topk=10):
        df = self.df.copy()
        df = self.demographic_filter(df, genres=genres, duration=duration, year=year)
        df = self.compute_imdb_score(df)
        result = df.sort_values('score', ascending=False).head(topk)
        result = result.loc[:, "title":"release_year"]
        return result
    @staticmethod
    def demographic_filter(df, genres=None, duration=None, year=None):
        df = df.copy()
        if genres:
            df = df[df[genres].all(axis=1)]
        if duration:
            df = df[df.runtime.between(duration[0], duration[1])]
        if year:
            df = df[df.release_year.between(year[0], year[1])]
        return df
    @staticmethod 
    def compute_imdb_score(df, q=0.85):
        df = df.copy()
        m = df.vote_count.quantile(q)
        C = (df.vote_average * df.vote_count).sum() / df.vote_count.sum()
        df = df[df.vote_count >= m]
        df['score'] = df.apply( lambda x: (x.vote_average * x.vote_count + C * m) / (x.vote_count + m), axis=1)
        return df
```

```python
recsys = RecommenderSystem(data="data/demographic.csv")
```

```python
recsys.recommend(genres=['Animation'], duration=(60, 150), year=(2010, 2019))
```
