```python
import pandas as pd
import googleapiclient.discovery
```

```python
# -*- coding: utf-8 -*-

# Sample Python code for youtube.commentThreads.list
# See instructions for running these code samples locally:
# https://developers.google.com/explorer-help/code-samples#python

# import os

def crawl(api_key, videoId, nextPageToken=None):


    api_service_name = "youtube"
    api_version = "v3"
    # print(nextPageToken)

    youtube = googleapiclient.discovery.build(api_service_name, api_version, developerKey = api_key)

    request = youtube.commentThreads().list(
        part="snippet,replies",
        textFormat="plainText",
        pageToken=nextPageToken,
        videoId=videoId
    )
    dataYoutube = request.execute()
    # return request.execute()
    # print(dataYoutube)
    hasil = []
    print(dataYoutube['items'])
    for data in dataYoutube['items']:
      hasil.append(data["snippet"]["topLevelComment"]["snippet"]["textOriginal"])
      if "replies" in data:
        for reply in data["replies"]["comments"]:
          hasil.append(reply["snippet"]["textOriginal"])

    # print(len(hasil))
    if "nextPageToken" in dataYoutube:
      # print(hasil)
      return  hasil + crawl(api_key, videoId, dataYoutube["nextPageToken"])
    else:
      # print(hasil)
      return hasil


if __name__ == "__main__":
  API_KEY = "AIzaSyBqQjcWqhVcoWErmPbwPXvNAvoXAVUAB1M"
  videoId = "SOiWx6hm3T0"
  # comments = ["Comments"]
  dataYoutube = crawl(API_KEY, videoId)
  # comments += dataYoutube
  pd.DataFrame(dataYoutube).to_csv("comments.csv")

  # saveData(dataYoutube)
```
