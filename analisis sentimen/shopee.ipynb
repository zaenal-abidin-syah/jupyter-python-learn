```python
import string 
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset_shopee.csv", sep=",", encoding="ISO-8859-1")
data.head(15)
```

```python
# lowercase
data["Review"] = data["Review"].apply(lambda x: x.lower())
data.head(15)
```

```python
# remove tanda baca dan angka
import re

def cleansing(review):
    return re.sub("\s+", ' ', re.sub(r"[^a-z|\s]", '', review.strip(" ")))
data["Review"] = data["Review"].apply(cleansing)
```

```python
import nltk
nltk.download("punkt")
```

```python
from nltk.tokenize import word_tokenize

data["Review"] = data["Review"].apply(lambda x: word_tokenize(x))
data.head()
```

```python
normalizad_word_dict = {}



def normalize(review):
    return [normalizad_word_dict[term] if term in normalized_word_dict else for term in review]
    
data["Review"].apply(normalize)
```
